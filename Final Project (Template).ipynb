```python
# TODO 07 - which top 5 States have the least total sales between 15 April 2019 - 31 December 2019
```

```python
import numpy as np
df_between = df[(df['Order Date'] >= '2019-04-15') & (df['Order Date'] <= '2019-12-31')]
df_tail5 = df_between.groupby('State')['Sales'].sum().sort_values().head(5)
df_tail5
```

```python
# TODO 08 - what is the proportion of total sales (%) in West + Central in 2019 e.g. 25% 
```

```python
df_2019 = df[df['Order Date'].dt.year == 2019]
df_2019_west = df_2019[df_2019['Region'] == "West"]['Sales'].sum()
df_2019_central = df_2019[df_2019['Region'] == "Central"]['Sales'].sum()
```

```python
total_sum = df_2019['Sales'].sum()
df_prop = ((df_2019_west + df_2019_central) / total_sum) * 100
print("Sum of sales for the West region in 2019:", df_prop, "%")
```

```python
# TODO 09 - find top 10 popular products in terms of number of orders vs. total sales during 2019-2020
```

```python
df_between2 = df[(df['Order Date'] >= '2019') & (df['Order Date'] <= '2020')]
product_stats = df_between2.groupby('Product Name')[['Sales', 'Quantity']].agg({'Quantity': 'count', 'Sales': 'sum'})
product_stats['Popularity'] = product_stats['Quantity'] / product_stats['Sales']
top_10 = product_stats.sort_values(by='Popularity', ascending=False).head(10)
top_10
```
